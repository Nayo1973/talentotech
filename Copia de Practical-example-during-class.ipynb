````python
import numpy as np
import pandas as pd

# Number of records to generate
num_records = 600

# Generate names
value_n_list = [f'Person_{i+1}' for i in range(num_records)]

# Generate random weights (between 40 and 120 kg)
value_w_list = np.random.uniform(40, 120, num_records).round(1).tolist()

# Generate random heights (between 1.50 and 2.00 meters)
value_h_list = np.random.uniform(1.50, 2.00, num_records).round(2).tolist()

# Create a dictionary with the new data
data_dict_large = {
    "name": value_n_list,
    "weight": value_w_list,
    "height": value_h_list
}

# Turn the dictionary into a DataFrame, overwriting the existing df
df = pd.DataFrame(data_dict_large)

# Calculate BMI (weight / height^2)
df['BMI'] = df['weight'] / (df['height'] ** 2)

# Define a function to determine health condition based on BMI
def determine_health_condition(bmi):
    if 18.5 <= bmi <= 24.9:
        return 'Normal weight'
    elif bmi < 18.5:
        return 'Underweight'
    elif bmi >= 25.0 and bmi <= 29.9:
        return 'Overweight'
    else:
        return 'Obesity'

# Apply the function to create the new column
df['health_condition'] = df['BMI'].apply(determine_health_condition)

print(f"DataFrame now contains {len(df)} rows.")
# Display the first 5 rows and the last 5 rows to show the expanded data
display(df.head())
display(df.tail())
````

## Generate Data for New Table

### Subtask:
Define lists of common Colombian first names (male and female), last names, document types, marital statuses, and a list of common Colombian cities to be used for data generation. Also, prepare helper functions for generating unique document numbers, email addresses, and realistic ages.


**Reasoning**:
The subtask requires defining several lists of common Colombian data (names, document types, marital statuses, cities) and helper functions for generating unique document numbers, email addresses, and realistic ages. This code block will initialize these lists and functions.



````python
import random

# 1. Define nombres_masculinos
nombres_masculinos = [
    'Andrés', 'Santiago', 'Juan', 'Diego', 'Mateo', 'Miguel', 'Daniel', 'Carlos', 'David', 'Felipe'
]

# 2. Define nombres_femeninos
nombres_femeninos = [
    'Sofía', 'Valentina', 'Isabella', 'Valeria', 'Mariana', 'Camila', 'Gabriela', 'Luciana', 'María', 'Sara'
]

# 3. Define apellidos_colombianos
apellidos_colombianos = [
    'Rodríguez', 'González', 'García', 'Martínez', 'Restrepo', 'López', 'Hernández', 'Díaz', 'Torres', 'Pérez',
    'Ramírez', 'Sánchez', 'Gómez', 'Castro', 'Vargas', 'Silva', 'Rojas', 'Ospina', 'Muñoz', 'Moreno'
]

# 4. Define tipos_documento
tipos_documento = ['CC', 'TI', 'CE', 'PPT'] # Cédula de Ciudadanía, Tarjeta de Identidad, Cédula de Extranjería, Permiso por Protección Temporal

# 5. Define estados_civiles
estados_civiles = ['Soltero', 'Casado', 'Divorciado', 'Viudo', 'Unión Libre']

# 6. Define ciudades_colombianas
ciudades_colombianas = [
    'Bogotá', 'Medellín', 'Cali', 'Barranquilla', 'Cartagena', 'Cúcuta', 'Bucaramanga', 'Pereira', 'Manizales', 'Neiva'
]

# Set to store generated document numbers to ensure uniqueness
generated_document_numbers = set()

# 7. Define function to generate unique document number
def generar_documento_unico():
    while True:
        # Generate a random 8 to 10 digit number
        doc_number = str(random.randint(10000000, 9999999999))
        if doc_number not in generated_document_numbers:
            generated_document_numbers.add(doc_number)
            return doc_number

# 8. Define function to generate an email address
def generar_email(nombre, apellido):
    # Convert names to lowercase and remove spaces/special characters
    nombre_clean = ''.join(filter(str.isalnum, nombre)).lower()
    apellido_clean = ''.join(filter(str.isalnum, apellido)).lower()
    domains = ['example.com', 'mail.co', 'web.org']
    return f"{nombre_clean}.{apellido_clean}@{random.choice(domains)}"

# 9. Define function to generate a realistic age
def generar_edad(min_edad=18, max_edad=65):
    return random.randint(min_edad, max_edad)

print("Lists and helper functions for data generation defined.")
````

**Reasoning**:
Now that the lists and helper functions for data generation are defined, the next logical step is to use them to generate 100 records and construct the `df_colombia` DataFrame, including all specified columns: 'tipo_documento', 'numero_documento', 'nombre_completo', 'sexo', 'estado_civil', 'edad', 'nacionalidad', 'correo_electronico', 'direccion', and 'ciudad'. Finally, display the first 10 rows of the newly created DataFrame.



````python
import pandas as pd
import random

# Prepare an empty list to hold the records
records = []

for _ in range(100):
    sexo = random.choice(['Masculino', 'Femenino'])
    if sexo == 'Masculino':
        nombre = random.choice(nombres_masculinos)
    else:
        nombre = random.choice(nombres_femeninos)

    apellido1 = random.choice(apellidos_colombianos)
    apellido2 = random.choice(apellidos_colombianos)
    # Ensure full name is unique if possible, though not strictly required by the prompt for uniqueness for name_completo
    nombre_completo = f"{nombre} {apellido1} {apellido2}"

    record = {
        'tipo_documento': random.choice(tipos_documento),
        'numero_documento': generar_documento_unico(), # Ensure unique document numbers
        'nombre_completo': nombre_completo,
        'sexo': sexo,
        'estado_civil': random.choice(estados_civiles),
        'edad': generar_edad(),
        'nacionalidad': 'Colombian',
        'correo_electronico': generar_email(nombre, apellido1),
        'direccion': f"Carrera {random.randint(1, 100)} # {random.randint(1, 200)}-{random.randint(1, 99)}",
        'ciudad': random.choice(ciudades_colombianas)
    }
    records.append(record)

# Create the DataFrame
df_colombia = pd.DataFrame(records)

print(f"DataFrame 'df_colombia' created with {len(df_colombia)} records.")

# Display the first 10 rows
display(df_colombia.head(10))
````
